###Aula 69 - Understanding Set Theory

###Aula 70 - Set Methods

SET METHODS

add --- union --- revome

difference --- intersection --- discard

clear --- copy --- update

issubset --- issuperset ---  

Set methods are inherited from the built-in set class



Python creates special hashes for each element in the set in order to ensure that all elements remain unique. And for that special reason, you are not allowed to add any mutable objects to the set.

You cannot add lists, dictionaries or other sets to the set in Python.

```python
# Set - ADD method
# Adding elements

photo_sizes = {'1920x1080', '800x600'}
print(photo_sizes)

photo_sizes.add('1024x768')
print(photo_sizes)
```

```python
# Set - UNION method
# Merging sets
photo_sizes = {'1920x1080', '800x600'}
other_sizes = {'800x600', '1024x768',}

# (You can also use the OR operator |) para obter o mesmo
# resultado do mtodo union
all_sizes = photo_sizes.union(other_sizes)
print(all_sizes) # Retorna todos os elementos contidos nos dois
                 # conjuntos sem repetio de um elemento do conj

```

```python
# Set - INTERSECTION method
# Intersection of sets
photo_sizes = {'1920x1080', '800x600'}
other_sizes = {'800x600', '1024x768',}

# (You can also use the ANd operator &) para obter o mesmo
# resultado do mtodo intersection
all_sizes = photo_sizes.intersection(other_sizes)
print(all_sizes) # Retorna todos os elementos contidos nos dois
                 # conjuntos sem repetio de um elemento do conj
```

```python
# Set - ISSUBSET method
# One set is included in another set
nums = {10, 5, 35}
other_nums = {20, 5, 12, 10, 35}

res = nums.issubset(other_nums)

print(res) # All elements of the first set are included in the second set
```

###Aula 84 - Strategies to Prevent Object Mutation

```python
# When you copy one variable to another, you technically copy only reference
# to specific object in memory. And if that object is mutable, for example, list
# or dictionary or object created based on your user defined class. You will be
# able to modify that single object via any of the variables. And this change
# will be reflected on all variables which point to the same object in memory.

# OPO 1
# Uma estratgia para copiar um objeto  usar os built-in methods. Por exemplo,
# abaixo temos um dicionrio chamado info, para fazer uma cpia podemos usar
# o mtodo copy( ), com isso teremos dois objetos independentes.

info = {
    'name': 'Marshal',
    'is_instructor': True
}

info_copy = info.copy()

info_copy['reviews_qty'] = 5

print(info)
print(info_copy)

print(id(info))
print(id(info_copy))
```

```python
# OPO 1
# Neste exemplo o dicionrio info tem uma chave chamada 'reviews' e o valor 
# uma lista vazia (mutable). novamente fazemos uma cpia com o mtodo copy( )
# criando um novo objeto chamado info_copy. Na sequncia estamos inserindo
# um novo valor na lista reviews do dicionrio info_copy utilizando um append.
# Aps isso vemos que tanto info como info_copy possuem o elemento inserido na
# lista. Isso acontece porque listas so objetos mutveis em Python e a chave
# reviews aponta para um local especfico na memria onde a lista est alocada.
# Quando fazemos a cpia de um objeto para um novo a referncia da lista no muda.
# Ento a chave reviews aponta para o mesmo endereo de memria onde est armazenada
# a lista. Ento o mtodo copy( ) faz uma cpia chamada shallow copy, ou seja, faz
# uma cpia de primeiro nvel. Mas se h objetos mutveis como neste exemplo, as
# referncias dos objetos aninhados so copiados para o outro objeto.


info = {
    'name': 'Marshal',
    'is_instructor': True,
    'reviews': []
}

info_copy = info.copy()
info_copy['reviews'].append('Great course!')

print(info)

print(info_copy)
```

```python
# OPO 2
# Uma maneira de resolver a situao do exemplo anterior  usar o mtodo deepcopy()
# como mostrado abaixo.

from copy import deepcopy

info = {
    'name': 'Marshal',
    'is_instructor': True,
    'reviews': []
}

info_deepcopy = deepcopy(info)
info_deepcopy['reviews'].append('Greate Course!')

print(info)
print(info_deepcopy)
```

###Aula 101 - Practice - Gathering All Keyword Arguments into the **kwargs Dictionary

```python
def send_email(to, subject, *args, **kwargs):
  print(f'Sending an email to: {to}')
  print(f'Email subject: {subject}')

  if args:
    print('Additional recipients: ')
    for recipient in args:
      print(recipient)

  if kwargs:
    print('Additional details for the email: ')
    for key in list(kwargs):
      print(f'{key}: {kwargs[key]}')

print('=========================================')
send_email('test@test.com', 'Hello there')
print('=========================================')
send_email('test@test.com', 'Hello there', bcc='marshal@gmail.com', img='test.png')
print('=========================================')
send_email('test@test.com', 'Hello there', 'other@test.com', 'someone@test.com')
print('=========================================')
send_email('test@test.com', 'Hello there', 'other@test.com', 'someone@test.com', bcc='marshal@gmail.com', img='test.png')
print('=========================================')
```

###Aula 113 - Practice - Using the Global Keyword

```python
"""
O que aconteceu aqui  que o Python tetou encontrar a varivel counter dentro da funo
E caso voc queira realizar tal operao com tal varivel, Python precisa
tenha esta varivel definida dentro da funo. No sai e no
procura por uma varivel como counter fora da funo. Ento voc tem que adicionar
counter seja como parmetro da funo, por exemplo. E neste caso voc tem que passar
counter aqui na chamada da funo. E agora, se executar o cdigo, no vai ter erro.
E basicamente agora dentro da funo inc_counter incrementamos o valor do parmetro,
mas no modificamos o valor da varivel counter. Se imprimir o valor da varivel
counter ainda vou ter zero.
Lembre-se counter  immutable object e neste caso counter += 1 cria um novo objeto
e agora o parmetro counter da funo aponta para esse novo objeto porque no podemos
modifiar o valor da varivel counter.
"""

counter = 0

def inc_counter(counter):
  counter += 1


def dec_counter():
  counter -= 1


inc_counter(counter)
print(counter)
```

```python
"""
Mas queremos modificar o valor do objeto counter de fora da funo, mas queremos
fazer isso dentro da funo. Para fazermos isso usamos a palavar global dentro da
funo como mostrado abaixo.


"""

counter = 0

def inc_counter():
  global counter  # Isso indica que estou trabalhando com a varivel global dentro da funo.
  counter += 1


def dec_counter():
  global counter  # Isso indica que estou trabalhando com a varivel global dentro da funo.
  counter -= 1


inc_counter()
print(counter)

inc_counter()
print(counter)

inc_counter()
print(counter)
```
