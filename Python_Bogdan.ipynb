###Aula 69 - Understanding Set Theory

###Aula 70 - Set Methods

SET METHODS

add --- union --- revome

difference --- intersection --- discard

clear --- copy --- update

issubset --- issuperset ---  

Set methods are inherited from the built-in set class



Python creates special hashes for each element in the set in order to ensure that all elements remain unique. And for that special reason, you are not allowed to add any mutable objects to the set.

You cannot add lists, dictionaries or other sets to the set in Python.

```python
# Set - ADD method
# Adding elements

photo_sizes = {'1920x1080', '800x600'}
print(photo_sizes)

photo_sizes.add('1024x768')
print(photo_sizes)
```

```python
# Set - UNION method
# Merging sets
photo_sizes = {'1920x1080', '800x600'}
other_sizes = {'800x600', '1024x768',}

# (You can also use the OR operator |) para obter o mesmo
# resultado do método union
all_sizes = photo_sizes.union(other_sizes)
print(all_sizes) # Retorna todos os elementos contidos nos dois
                 # conjuntos sem repetição de um elemento do conj

```

```python
# Set - INTERSECTION method
# Intersection of sets
photo_sizes = {'1920x1080', '800x600'}
other_sizes = {'800x600', '1024x768',}

# (You can also use the ANd operator &) para obter o mesmo
# resultado do método intersection
all_sizes = photo_sizes.intersection(other_sizes)
print(all_sizes) # Retorna todos os elementos contidos nos dois
                 # conjuntos sem repetição de um elemento do conj
```

```python
# Set - ISSUBSET method
# One set is included in another set
nums = {10, 5, 35}
other_nums = {20, 5, 12, 10, 35}

res = nums.issubset(other_nums)

print(res) # All elements of the first set are included in the second set
```

###Aula 83 - Understanding Mutable Objects in Python

```python
# Ids of the mutable objects
# Os endereços na memória são diferentes
# Objects in memory are different, and that's what Python does for all mutable
# objects. It does not reuse existing objects. It creates brand new objects in
# case you create, for example, list. Ele não copiou my_list em other_list ele
# uma nova lista. No exemplo abaixo ele criou 3 novas listas.

my_list = [1, 2, 3]
print(id(my_list))

other_list = [1, 2, 3]
print(id(other_list))

print(id([1, 2, 3]))
```

###Aula 84 - Strategies to Prevent Object Mutation

```python
# When you copy one variable to another, you technically copy only reference
# to specific object in memory. And if that object is mutable, for example, list
# or dictionary or object created based on your user defined class. You will be
# able to modify that single object via any of the variables. And this change
# will be reflected on all variables which point to the same object in memory.

# OPÇÃO 1
# Uma estratégia para copiar um objeto é usar os built-in methods. Por exemplo,
# abaixo temos um dicionário chamado info, para fazer uma cópia podemos usar
# o método copy( ), com isso teremos dois objetos independentes.

info = {
    'name': 'Marshal',
    'is_instructor': True
}

info_copy = info.copy()

info_copy['reviews_qty'] = 5

print(info)
print(info_copy)

print(id(info))
print(id(info_copy))
```

```python
# OPÇÃO 1
# Neste exemplo o dicionário info tem uma chave chamada 'reviews' e o valor é
# uma lista vazia (mutable). novamente fazemos uma cópia com o método copy( )
# criando um novo objeto chamado info_copy. Na sequência estamos inserindo
# um novo valor na lista reviews do dicionário info_copy utilizando um append.
# Após isso vemos que tanto info como info_copy possuem o elemento inserido na
# lista. Isso acontece porque listas são objetos mutáveis em Python e a chave
# reviews aponta para um local específico na memória onde a lista está alocada.
# Quando fazemos a cópia de um objeto para um novo a referência da lista não muda.
# Então a chave reviews aponta para o mesmo endereço de memória onde está armazenada
# a lista. Então o método copy( ) faz uma cópia chamada shallow copy, ou seja, faz
# uma cópia de primeiro nível. Mas se há objetos mutáveis como neste exemplo, as
# referências dos objetos aninhados são copiados para o outro objeto.


info = {
    'name': 'Marshal',
    'is_instructor': True,
    'reviews': []
}

info_copy = info.copy()
info_copy['reviews'].append('Great course!')

print(info)

print(info_copy)
```

```python
# OPÇÃO 2
# Uma maneira de resolver a situação do exemplo anterior é usar o método deepcopy()
# como mostrado abaixo.

from copy import deepcopy

info = {
    'name': 'Marshal',
    'is_instructor': True,
    'reviews': []
}

info_deepcopy = deepcopy(info)
info_deepcopy['reviews'].append('Greate Course!')

print(info)
print(info_deepcopy)
```

###Aula 101 - Practice - Gathering All Keyword Arguments into the **kwargs Dictionary

```python
def send_email(to, subject, *args, **kwargs):
  print(f'Sending an email to: {to}')
  print(f'Email subject: {subject}')

  if args:
    print('Additional recipients: ')
    for recipient in args:
      print(recipient)

  if kwargs:
    print('Additional details for the email: ')
    for key in list(kwargs):
      print(f'{key}: {kwargs[key]}')

print('=========================================')
send_email('test@test.com', 'Hello there')
print('=========================================')
send_email('test@test.com', 'Hello there', bcc='marshal@gmail.com', img='test.png')
print('=========================================')
send_email('test@test.com', 'Hello there', 'other@test.com', 'someone@test.com')
print('=========================================')
send_email('test@test.com', 'Hello there', 'other@test.com', 'someone@test.com', bcc='marshal@gmail.com', img='test.png')
print('=========================================')
```

###Aula 113 - Practice - Using the Global Keyword

```python
"""
O que aconteceu aqui é que o Python tetou encontrar a variável counter dentro da função
E caso você queira realizar tal operação com tal variável, Python precisa
tenha esta variável definida dentro da função. Não sai e não
procura por uma variável como counter fora da função. Então você tem que adicionar
counter seja como parâmetro da função, por exemplo. E neste caso você tem que passar
counter aqui na chamada da função. E agora, se executar o código, não vai ter erro.
E basicamente agora dentro da função inc_counter incrementamos o valor do parâmetro,
mas não modificamos o valor da variável counter. Se imprimir o valor da variável
counter ainda vou ter zero.
Lembre-se counter é immutable object e neste caso counter += 1 cria um novo objeto
e agora o parâmetro counter da função aponta para esse novo objeto porque não podemos
modifiar o valor da variável counter.
"""

counter = 0

def inc_counter(counter):
  counter += 1


def dec_counter():
  counter -= 1


inc_counter(counter)
print(counter)
```

```python
"""
Mas queremos modificar o valor do objeto counter de fora da função, mas queremos
fazer isso dentro da função. Para fazermos isso usamos a palavar global dentro da
função como mostrado abaixo.


"""

counter = 0

def inc_counter():
  global counter  # Isso indica que estou trabalhando com a variável global dentro da função.
  counter += 1


def dec_counter():
  global counter  # Isso indica que estou trabalhando com a variável global dentro da função.
  counter -= 1


inc_counter()
print(counter)

inc_counter()
print(counter)

inc_counter()
print(counter)
```
